```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.myconfig import Config
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib
import random
config = Config()
random.seed(0)
plt.style.use(config.style)
indexes={'舒适':0,'耐久':0,'精细':0,'保障':0,'牌证':0,'外观':0,'外部':0,'响应及时性':0,'沟通有效性':0,'信息安全性':0,'补偿合理性':0,'服务专业性':0}
# 分类的方法
data=pd.read_csv('D:\python_common\代码_感知质量大论文\公开数据处理\感知质量数据集\随机占优PROMETHEE-II过程\E出入流汇总.csv',encoding='utf-8',index_col=0)
data.index=['ALBD','AND','AT','BNL','GRN','LN','QPL','TB']
style=['--o','--s','--p','--h','--^','--*','--+','--x']
data=data.T.sort_values(by='ALBD',ascending=False)

plt.figure(figsize=(15,7),dpi=config.dpi)
for i in range(0,len(data.columns)):
    plt.plot(data.index,data.iloc[:,i],style[i],label=data.columns[i],linewidth=2,markersize=8)

plt.xticks(range(0,len(data.index)),data.index,rotation=config.rotation,fontsize=config.ticks_fontsize,fontfamily=config.chinese_fontfamily)
plt.yticks(fontsize=config.ticks_fontsize,fontfamily=config.chinese_fontfamily)
plt.ylabel('净流量',fontsize=config.label_fontsize,fontfamily=config.chinese_fontfamily)
plt.xlabel('维度',fontsize=config.label_fontsize,fontfamily=config.chinese_fontfamily)

config.invisible_up_right(plt)
config.my_legend(plt,kwargs={'prop':{'size':17,'family':'times new roman'}})
plt.show()
```
