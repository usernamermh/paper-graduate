```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# 
plt.rcParams['axes.grid'] = False
plt.figure(figsize=(10,3), dpi=220)
plt.rcParams['font.weight'] = 'bold'

data=pd.read_excel('/.xls',index_col=0)

import scipy.cluster.hierarchy as sch # 

dendrogram = sch.dendrogram(
    sch.linkage(data.to_numpy() , method = 'ward'),
    labels=data.index,
    leaf_font_size=10,
    leaf_rotation=0,
    orientation='top',
    color_threshold=0,
    above_threshold_color='black',
    distance_sort='descending',
    show_leaf_counts=True,


) # 
#
plt.xlabel('',fontsize=13,fontfamily='STsong',fontweight='bold',color='black')
plt.ylabel('\n',fontsize=13,fontfamily='STsong',fontweight='bold',rotation=0,verticalalignment='center',labelpad=5,rotation_mode='anchor',color='black')

ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.yticks(fontsize=7,fontfamily='Times New Roman',fontweight='bold')


# 
labels2 = []
labels = ax.get_xticklabels()
for each in labels:
    text=each.get_text()
    text=text[0]+'\n'+text[1]
    labels2.append(text)
ax.set_xticklabels(labels2,fontsize=8,fontfamily='STsong',
                   fontweight='bold',
                   verticalalignment='top')

plt.show()
```
