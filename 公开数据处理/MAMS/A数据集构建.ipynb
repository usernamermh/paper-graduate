```python
from xml.etree.ElementTree import parse
import re
# 不处理test，拆分train
path='test'
data1 = parse(path+'.xml')
root=data1.getroot()
sentiment2id = {'negative':'1','neutral':'2','positive':'3'}

def get_text(root):
    with open(path+'.tsv','w',encoding='gbk') as f:
        f.write('labels\tkey_words\treview\n')
        review = root.findall('sentence')
        for each in review:
            text = each.find('text').text
            aspectcategories = each.find('aspectCategories')
            for each2 in aspectcategories:
                category = each2.get('category')
                polarity = each2.get('polarity')
                f.write(sentiment2id[polarity]+category+'\t'+category+'\t'+text+'\n')
get_text(root)
```

```python
import pandas as pd
import numpy as np
ALL_cats=[]
path1='test'
path2='train' if path1=='test' else 'test'
path3='dev'

data1=pd.read_csv(path1+'.tsv',delimiter='\t')
data1['senti']=data1['labels'].apply(lambda x:x[0])
data1['cat']=data1['labels'].apply(lambda x:x[1:])
data1=data1[['senti','cat','key_words','review']]
data1_=pd.DataFrame(columns=['CAT','KEYWORDS','review'])
for i in range(len(data1)):
    review=data1['review'][i]
    cat=data1['cat'][i]
    ALL_cats.append(cat)
    senti=data1['senti'][i]
    keywords=data1['key_words'][i] if type(data1['key_words'][i])==str else 'NEG'
    data1_=data1_.append(
        pd.DataFrame(
            [[cat,str(senti)+' '+keywords,review]],
            columns=['CAT','KEYWORDS','review']
        )
    )

data2=pd.read_csv(path2+'.tsv',delimiter='\t')
data2['senti']=data2['labels'].apply(lambda x:x[0])
data2['cat']=data2['labels'].apply(lambda x:x[1:])
data2=data2[['senti','cat','key_words','review']]
data2_=pd.DataFrame(columns=['CAT','KEYWORDS','review'])
for i in range(len(data2)):
    review=data2['review'][i]
    cat=data2['cat'][i]
    ALL_cats.append(cat)
    senti=data2['senti'][i]
    keywords=data2['key_words'][i] if type(data2['key_words'][i])==str else 'NEG'
    data2_=data2_.append(
        pd.DataFrame(
            [[cat,str(senti)+' '+keywords,review]],
            columns=['CAT','KEYWORDS','review']
        )
    )

data3=pd.read_csv(path3+'.tsv',delimiter='\t')
data3['senti']=data3['labels'].apply(lambda x:x[0])
data3['cat']=data3['labels'].apply(lambda x:x[1:])
data3=data3[['senti','cat','key_words','review']]
data3_=pd.DataFrame(columns=['CAT','KEYWORDS','review'])
for i in range(len(data3)):
    review=data3['review'][i]
    cat=data3['cat'][i]
    ALL_cats.append(cat)
    senti=data3['senti'][i]
    keywords=data3['key_words'][i] if type(data3['key_words'][i])==str else 'NEG'
    data3_=data3_.append(
        pd.DataFrame(
            [[cat,str(senti)+' '+keywords,review]],
            columns=['CAT','KEYWORDS','review']
        )
    )

classes=list(set(ALL_cats))
# 排序
for each in classes:
    try:
        print(each)
    except:
        pass
l = ' '.join(classes)
with open('label.txt','w') as f:
    for each in classes:
        f.write(each+'\n')
data1_['KEYWORDS'] = data1_['KEYWORDS'].apply(lambda x:x[0]+' '+l)
data2_['KEYWORDS'] = data2_['KEYWORDS'].apply(lambda x:x[0]+' '+l)
data3_['KEYWORDS'] = data3_['KEYWORDS'].apply(lambda x:x[0]+' '+l)
data1_.to_csv('关键词抽取结果_'+path1+'.csv',index=False,header=['labels','key_words','review'],encoding='utf-8')
data2_.to_csv('关键词抽取结果_'+path2+'.csv',index=False,header=['labels','key_words','review'],encoding='utf-8')
data3_.to_csv('关键词抽取结果_'+path3+'.csv',index=False,header=['labels','key_words','review'],encoding='utf-8')
```
