```python
import os
import matplotlib.pyplot as plt
from matplotlib.myconfig import  Config
from collections import Counter
config=Config()
config()
plt.style.use('mpl20')
with open('分年份数据.txt','r',encoding='utf-8') as f:
    data=f.readlines()
Times=[]
for each in data:
    times=each[-20:-16]
    if '2019' in times or '2023-07' in times or '2023-08' in times:
        continue
    elif times.startswith('5'):
        times=times[1:]+'0'
    else:
        Times.append(times)

# 分月份计数
c=Counter(Times)
c=sorted(c.items(),key=lambda x:x[0])
# 计算百分比
sum=0
for each in c:
    sum+=each[1]
for i in range(len(c)):
    c[i]=(c[i][0],c[i][1]/sum)

plt.rcParams['font.sans-serif']=config.chinese_fontfamily
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=config.rectangle_shape,dpi=config.dpi)
plt.bar([each[0] for each in c],[each[1] for each in c],color='white',edgecolor='black',linewidth=1.5,linestyle='-',width=0.5,align='center',zorder=2)

plt.xticks(fontsize=config.ticks_fontsize)
plt.yticks([0,0.05,0.1,0.15,0.2,0.25,0.3,0.35],['0%', '5%', '10%', '15%', '20%', '25%', '30%', '35%'], fontsize=config.ticks_fontsize)
plt.xlabel('年份',fontsize=config.label_fontsize)
plt.ylabel('占比',fontsize=config.label_fontsize,labelpad=9.5)
plt.show()

print(c)
```

```python
import os
import matplotlib.pyplot as plt
from matplotlib.myconfig import  Config
from collections import Counter
import numpy as np
config=Config()
# config()
plt.style.use('mpl20')
# 读取所有的评论
data_all = []
for each in os.listdir('D:\python_common\代码_感知质量大论文\公开数据处理\感知质量数据集\爬取结果'):
    data = open('D:\python_common\代码_感知质量大论文\公开数据处理\感知质量数据集\爬取结果\\'+each,encoding='utf-8').readlines()
    data_all.extend(data)

dictt= {}
for each in data_all:
    if len(each) not in dictt:
        dictt.update({len(each):1})
    else:
        dictt[len(each)] +=1
dictt=sorted(dictt.items(),reverse=True,key=lambda x:x[0])

plt.figure(figsize=config.rectangle_shape,dpi=config.dpi)

X=[each[0] for each in dictt]
Y=[each[1] for each in dictt]
print(max(X))

# 箱线图
f = plt.boxplot(X,showfliers=False)
plt.xticks([])
plt.yticks(fontsize=config.ticks_fontsize,fontfamily=config.english_fontfamily)
plt.ylabel('评论长度',fontsize=config.label_fontsize,fontfamily=config.chinese_fontfamily)

# 线的颜色
for each in f['medians']:
    each.set(c='black')

# 计算下四分位数
Q1 = np.percentile(X, 25)
# 计算上四分位数
Q2 = np.percentile(X, 75)

plt.hlines(Q1,0.5,1.5,colors='black',linestyles='--')
plt.hlines(Q2,0.5,1.5,colors='black',linestyles='--')
plt.xlim((0.5,1.5))

plt.text(0.52,Q1-45,str(Q1)[0:2],fontsize=config.ticks_fontsize,fontfamily=config.english_fontfamily)
plt.text(0.52,Q2+15,str(Q2)[0:3],fontsize=config.ticks_fontsize,fontfamily=config.english_fontfamily)


plt.show()
plt.close('all')
```
